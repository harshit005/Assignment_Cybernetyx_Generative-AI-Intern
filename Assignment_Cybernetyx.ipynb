```python
import nest_asyncio
import uvicorn
from fastapi import FastAPI, UploadFile, File, HTTPException
from sentence_transformers import SentenceTransformer
import chromadb
from chromadb.config import Settings
import asyncio

# Enable async support in Jupyter Notebook
nest_asyncio.apply()
```

```python
# Initialize ChromaDB with persistent storage
chroma_client = chromadb.Client(Settings())

# Initialize the Sentence-Transformers model for embeddings
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
app = FastAPI()

# Helper function to get or create a collection in ChromaDB
def get_or_create_collection(name="documents"):
    if name not in chroma_client.list_collections():
        collection = chroma_client.create_collection(name)
    else:
        collection = chroma_client.get_collection(name)
    return collection

# Async function to process files and add to ChromaDB
async def process_and_store_file(file: UploadFile):
    try:
        content = (await file.read()).decode("utf-8")
        embedding = model.encode(content, convert_to_tensor=False).tolist()

        # Use the filename as the document ID
        document_id = file.filename

        # Get or create collection and add document
        collection = get_or_create_collection()
        collection.add(ids=[document_id], embeddings=[embedding], documents=[content])
    except Exception as e:
        raise HTTPException(status_code=500, detail=f"Failed to process file: {str(e)}")
```

```python
@app.post("/ingest")
async def ingest(file: UploadFile = File(...)):
    await process_and_store_file(file)
    return {"status": "Document ingested successfully", "filename": file.filename}

@app.get("/query")
async def query_documents(query: str):
    try:
        # Generate embedding for the query
        query_embedding = model.encode(query, convert_to_tensor=False).tolist()

        # Retrieve similar documents from ChromaDB
        collection = get_or_create_collection()
        results = collection.query(query_embedding=query_embedding, n_results=5)

        # Return top matches
        return {"query": query, "matches": results}
    except Exception as e:
        raise HTTPException(status_code=500, detail=f"Query failed: {str(e)}")
```

```python
# Run the FastAPI server (within the notebook environment)
def run_server():
    uvicorn.run(app, host="0.0.0.0", port=8000)

# Execute server
run_server()
```
